After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
1. title
2. Ratings
3. Price

```python
import selenium 
import pandas as pd
from selenium import webdriver
```

```python
# connecting webdriver
driver = webdriver.Chrome(r"C:\Users\tusha\Downloads\chromedriver.exe")
```

```python
url = 'https://www.amazon.in/'
driver.get(url)
```

```python
search=driver.find_element_by_xpath('//*[@id="twotabsearchtextbox"]')
search.send_keys('Laptop')
```

```python
search_btn=driver.find_element_by_xpath('//*[@id="nav-search-bar-form"]/div[3]').click()
```

```python
cpu1=driver.find_element_by_xpath('//*[@id="p_n_feature_thirteen_browse-bin/12598163031"]/span/a/span').click()
```

```python
cpu2=driver.find_element_by_xpath('//*[@id="p_n_feature_thirteen_browse-bin/16757432031"]/span/a/span').click()
```

```python
title=driver.find_elements_by_xpath('//div[@class="a-section a-spacing-none"]')
title
```

```python
titles=[]
for i in title[:15]:
    titles.append(i.text)
titles
```

```python
len(titles)
```

```python
price=driver.find_elements_by_xpath('//span[@class="a-price-whole"]')
price
```

```python
prices=[]
for i in price[:15]:
    prices.append(i.text)
prices
```

```python
len(prices)
```

```python
rating=driver.find_elements_by_xpath('//*[@id="a-popover-content-3"]/div/div/div/div[1]/span')
rating
```

```python
ratings_urls = []

url = driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')

for i in url:
    ratings_urls.append(i.get_attribute('href'))
ratings_urls
```

```python
ratings = []
for i in ratings_urls[:20]:
    driver.get(i)
    try:
        rating=driver.find_element_by_xpath('//*[@id="reviewsMedley"]/div/div[1]/div[2]/div[1]/div/div[2]/div/span/span')
        ratings.append(rating.text)
    except:
        pass
ratings
```

```python
len(ratings)
```

```python
ratings.pop(15)
```

```python
Laptops=pd.DataFrame({})
Laptops['Title']=titles
Laptops['Rating']=ratings
Laptops['Price']=prices
```
