# Find whether a number is prime or composite.

```python
num = int(input("Enter any number : "))
if num > 1:
    for i in range(2,num):
        if (num % i) == 0:
            print(num, "is NOT a prime number")
            break
    else:
        print(num, "is a PRIME number")
elif num == 0 or 1:
    print(num, "is a neither prime NOR composite number")
else:
    print(num, "is NOT a prime number it is a COMPOSITE number")
```

# Check whether a given string is palindrome or not.

```python
string = str(input("Enter any string "))
rev_str = reversed(string)
if list(string) == list(rev_str):
    print("The string is a palindrome.")
else:
    print("This string is not a palindrome.")
```

# To get the third side of right-angled triangle from two given side.

```python
base = float(input("Enter the base of a right-angled triangle: "))
height = float(input("Enter the height of a right-angled trinagle: "))

def findHypotenuse(base, height):
    
    h = (((base*base)+(height*height))**(1/2));
    return h;

print("The hypotenuse of a right-angled triangle is:")

print(findHypotenuse(base, height))
```

# To print the frequency of each of the characters present in a given string.

```python
string=input("Enter the string ")
freq=[None]*len(string)
for i in range(0,len(string)):
  freq[i]=1
  for j in range(i+1,len(string)):
    if(string[i]==string[j]):
        freq[i]=freq[i]+1
        string=string[:j]+'0'+string[j+1:];
print("Character and their frequency");
for i in range(0,len(freq)):
    if(string[i]!=' ' and string[i]!='0'):
        print(string[i]+"="+str(freq[i]))
```
